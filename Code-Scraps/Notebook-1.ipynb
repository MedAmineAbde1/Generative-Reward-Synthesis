```python
import math as m
from math import inf
import numpy as np

class Box_in_Chessboard :
    def __init__(self,name,xy:tuple[int,int], color:str, symbol:str=None, dirlist = list):
        # this part describes the box itself
        self.name=name
        self.coordinate=xy
        self.color=color
        self.reward=0
        self.symbol= symbol
        #this one is about the surrounding of the box
        self.neighbours={
            "up":dirlist[0],
            "down":dirlist[1],
            "right":dirlist[2],
            "left":dirlist[3],
            "stand":self
        }

    def reward_(self):
        if self.symbol == "X": self.reward = -100
        elif self.symbol == "O" : self.reward = 100 #m.inf
        elif self.name == "emptybox" : self.reward = -10
        else : self.reward = 0
    def next_state(self,action):
        if self.neighbours[action] is None:
            return Box_in_Chessboard("emptybox",(-1 ,-1),"No color","empty",[ 0 for i in range(4)])
        return self.neighbours[action]
    
#define the methode emptybox to limit the range of our box 
# class EmptyBox(Box_in_Chessboard):
#     def __init__(self, name, xy, color=None, symbol = None, dirlist=list):
#         super().__init__(name, xy, color, symbol, dirlist)


    
```

```python
#Computing Q_pi(s,a):
def Q_pi(state:Box_in_Chessboard, action,pi, gamma =0.1):
    #
    next_state = state.next_state(action)
    
    next_state.reward_() #compute reward of reaching the state
    r =  next_state.reward  #store the reward attribute within the r variable

    return r + gamma*V_pi(next_state , pi)

```

```python
# In this cell, it is thought to be compute pi* the optimal policy :

pi_star={ # state : action
} 

def pi_star_func(States: list, Actions: list, gamma: float = 0.9) -> dict:

    # ---- Phase 1 : greedy initialization ----
    # for each state, pick the action that maximizes immediate reward
    # this gives us a starting policy built from scratch (no prior pi)

    pi = {}
    state_action_recorder = {}

    for state in States:

        if state.symbol in ("X", "O"):  # terminal states, no decision needed
            pi[state] = "stand"
            state_action_recorder[state] = {"stand": state.reward}
            continue

        action_recorder = {}
        for action in Actions:
            next_state = state.next_state(action)
            next_state.reward_()
            action_recorder[action] = next_state.reward

        state_action_recorder[state] = action_recorder
        pi[state] = max(action_recorder, key=action_recorder.get)
        print(f"[Phase 1] {state.name} → best immediate action: {pi[state]} | r = {action_recorder[pi[state]]}")

    # ---- Phase 2 : Q-based verification and correction ----
    # now that we have a full policy, re-evaluate each state using Q_pi
    # which accounts for future rewards via gamma * V_pi(s')
    # we build pi_verified separately to avoid corrupting the pi used by Q_pi

    pi_verified = {}

    for state in States:

        if state.symbol in ("X", "O"):
            pi_verified[state] = pi[state]
            continue

        # sort actions best-to-worst by immediate reward (phase 1 data)
        sorted_actions = sorted(
            state_action_recorder[state],
            key=state_action_recorder[state].get,
            reverse=True  # best first
        )

        # Q value of the greedily chosen action under current pi
        best_action = pi[state]
        best_q = Q_pi(state, best_action, pi, gamma)

        # check all alternatives against it
        for action in sorted_actions:
            if action == best_action:
                continue  # already computed

            alt_q = Q_pi(state, action, pi, gamma)
            print(f"[Phase 2] {state.name} | {action}: Q={alt_q:.3f} vs current best {best_action}: Q={best_q:.3f}")

            if alt_q > best_q:
                best_q = alt_q
                best_action = action

        pi_verified[state] = best_action
        print(f"[Phase 2] {state.name} → final action: {best_action}\n")

    return pi_verified


pi_star = pi_star_func(States=States, Actions=Actions, gamma=0.9)

for state, action in pi_star.items():
    print(f"{state.name} → {action}")
```

```python
def pi_star_optimized(States,Actions,gamma=0.9): 
    
    # ---- Phase 1 : greedy initialization ----
    # for each state, pick the action that maximizes immediate reward
    # this gives us a starting policy built from scratch (no prior pi)

    pi = {}
    state_action_recorder = {}

    for state in States:

        if state.symbol in ("X", "O"):  # terminal states, no decision needed
            pi[state] = "stand"
            state_action_recorder[state] = {"stand": state.reward}
            continue

        action_recorder = {}
        for action in Actions:
            next_state = state.next_state(action)
            next_state.reward_()
            action_recorder[action] = next_state.reward

        state_action_recorder[state] = action_recorder
        pi[state] = max(action_recorder, key=action_recorder.get)
        print(f"[Phase 1] {state.name} → best immediate action: {pi[state]} | r = {action_recorder[pi[state]]}")

    # ---- Phase 2 : Q-based verification and correction ----
    # ---- Outer loop: repeat until policy stable ----
    
    iteration=0
    while True:
        changed = False
        iteration +=1
        pi_new={}

        for state in States:
            if state.symbol in ("X", "O"):
                pi_new[state] = pi[state]
                continue

            best_action = pi[state]
            best_q = Q_pi(state, best_action, pi, gamma)

            for action in Actions:
                if action == best_action:
                    continue  # already computed

                alt_q = Q_pi(state, action, pi, gamma)

                if alt_q > best_q:
                    best_q = alt_q
                    best_action = action
            
            pi_new[state]=best_action
            if best_action != pi[state]:
                changed = True
            
        pi = pi_new

        print(f"[Iteration {iteration}] policy changed: {changed}")

        if not changed: break # policy is stable → this is π*
    
    return pi

```

```python
import numpy as np

def V_Iteration(States: list, Actions: list, T: int = 4, gamma: float = 0.9):
    """
    Value Iteration Algorithm:
    V_{k+1}(s) = max_a [ r(s,a) + gamma * V_k(s') ]  for all s in S
    Returns V_T (final values) and pi* (optimal policy dict)
    """
    N = len(States)
    M = len(Actions)

    # V_array[k][n] = value of state n at iteration k
    V_array    = np.zeros((T, N))

    # policy_array[k][n] = best action for state n at iteration k
    policy_array = np.full((T, N), None, dtype=object)

    for k in range(T - 1):
        for n in range(N):

            # --- terminal states ---
            if States[n].symbol in ("X", "O"):
                States[n].reward_()
                V_array[k + 1][n]    = States[n].reward
                policy_array[k + 1][n] = "stand"
                continue

            # --- evaluate each action ---
            temp_values  = [-m.inf] * M
            temp_actions = [None]   * M

            for i, action in enumerate(Actions):
                next_state = States[n].next_state(action)

                if next_state.name == "emptybox":
                    continue

                next_state.reward_()
                n_ = States.index(next_state)

                temp_values[i]  = next_state.reward + gamma * V_array[k][n_]
                temp_actions[i] = action

            # --- take the max ---
            best_i = int(np.argmax(temp_values))

            V_array[k + 1][n]      = temp_values[best_i]
            policy_array[k + 1][n] = temp_actions[best_i]

    # --- build final policy as a dict {state: best_action} ---
    pi_star = {}
    for n, state in enumerate(States):
        pi_star[state] = policy_array[T - 1][n]

    return V_array[T - 1], pi_star


# --- run ---
T      = 9
values, pi_star = V_Iteration(States, Actions, T=T, gamma=0.9)

# --- display ---
print("=== Value Iteration Results ===\n")
for i, state in enumerate(States):
    print(f"  {state.name} | V* = {values[i]:>10.3f} | π* = {pi_star[state]}")
```

```python
import random
import math

def Policy_Iteration(States: list, Actions: list, T: int = 100, gamma: float = 0.9):
    pi = {s: random.choice(Actions) for s in States}

    for k in range(T - 1):
        V_pi = Policy_Evaluation0(States, Actions, Policy=pi, T=T, gamma=gamma)
        pi_new = {}

        for n, s in enumerate(States):
            temp_buffer = [(-math.inf, None)] * len(Actions)  # ← use math.inf directly

            for i, action in enumerate(Actions):              # ← i instead of m
                next_state = s.next_state(action)
                if next_state.name == "emptybox":
                    continue
                next_state.reward_()
                next_index = States.index(next_state)
                expected_value = next_state.reward + gamma * V_pi[next_index]
                temp_buffer[i] = (expected_value, action)     # ← i instead of m

            best_action = max(temp_buffer, key=lambda x: x[0])[1]
            pi_new[s] = best_action if best_action is not None else pi[s]

        pi = pi_new

    return V_pi, pi


V, pi_iteration = Policy_Iteration(States, Actions)
for s, a in pi_iteration.items():
    print(f"  {s.name} → {a}")
```
